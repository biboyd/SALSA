# Example Guide to using SALSA to extract absorbers from a dataset

First we need to download a dataset to work with. Here we'll use a yt test dataset of an Isolated Galaxy Simulation

```python
import yt
import salsa
import numpy as np
from astropy.units import kpc
```

```python
# Download one of yt's sample datasets
# Requires the "pooch" package
ds = yt.load_sample("IsolatedGalaxy")

raydir = './'

nrays = 2

ion_list = ['H I', 'C IV']  # A list of ions for which to look for absorbers

center=[0.5, 0.5, 0.5],  # Center from which to measure impact paramter, in code units
impact_range = [0, 50]  # Range of impact parameters. Defaults to kpc if nothing is specified.
ray_length = 200 * kpc  # Units can be specified with astropy or unyt.

add_fields = ['temperature', 'metallicity']  # Additional fields to extract & save in the dataset
field_units = {'metallicity':'Zsun'}  # Units to convert those additional fields to

np.random.seed(567)
tab = salsa.generate_catalog(ds, nrays, raydir, ion_list,
                             method='spice',  # currently the only supported method
                             center = center,
                             impact_param_lims = impact_range,
                             ray_length = ray_length,
                             fields = add_fields,
                             extractor_kwargs={'velocity_res':25, 'absorber_min':12.8, 'frac':0.8},
                             units_dict = field_units)
```
